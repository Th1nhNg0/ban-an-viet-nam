```python
import pandas as pd
```

```python
df = pd.read_json("./output/banan.jsonl", lines=True)
df.head()
```

```python
df.rename(
    columns={
        "Tên bản án:": "Ten_ban_an",
        "Quan hệ pháp luật:": "Quan_he_phap_luat",
        "Cấp xét xử:": "Cap_xet_xu",
        "Loại vụ/việc:": "Loai_vu_viec",
        "Tòa án xét xử:": "Toa_an_xet_xu",
        "Áp dụng án lệ:": "Ap_dung_an_le",
        "Đính chính:": "Dinh_chinh",
        "Thông tin về vụ/việc:": "Thong_tin_ve_vu_viec",
        "Đối tượng khởi kiện:": "Doi_tuong_khoi_kien",
        "Loại án:": "Loai_an",
        "Thông tin về vụ án:": "Thong_tin_ve_vu_an",
        "Tên quyết định:": "Ten_quyet_dinh",
        "Biện pháp xử lý hành chính:": "Bien_phap_xu_ly_hanh_chinh",
        "Loại việc:": "Loai_viec",
        "Thông tin về việc:": "Thong_tin_ve_viec",
        "Đối tượng bị yêu cầu:": "Doi_tuong_bi_yeu_cau",
        "Loại yêu cầu:": "Loai_yeu_cau",
        "Thông tin về yêu cầu:": "Thong_tin_ve_yeu_cau",
    },
    inplace=True,
)
```

```python
hinhsu_df = df[(df["Loai_an"] == "Hình sự") & (~df['Ten_ban_an'].isnull())]
```

```python
urls = hinhsu_df['url'].tolist()
```

```python
# export to txt
with open('hinhsu_urls.txt', 'w') as f:
    for url in urls:
        download_url = url.replace('2ta','5ta')
        file_name = download_url.split('/')[-2]
        f.write(f"{download_url}, {file_name}.pdf\n")
```

```python
# drop columns
hinhsu_df.drop(
    columns=[
        "Quan_he_phap_luat",
        "Loai_vu_viec",
        "Thong_tin_ve_vu_viec",
        "Doi_tuong_khoi_kien",
        "Ten_quyet_dinh",
        "Bien_phap_xu_ly_hanh_chinh",
        "Loai_viec",
        "Thong_tin_ve_viec",
        "Doi_tuong_bi_yeu_cau",
        "Loai_yeu_cau",
        'Thong_tin_ve_yeu_cau'
    ],
    inplace=True,
)
```
